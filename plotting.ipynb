```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# ===============================
# Helper Functions for Plotting
# ===============================

def plot_histogram(data, labels, title, xlabel, ylabel, ax=None):
    """Plots a histogram using a bar chart."""
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 5))

    ax.bar(labels, data, color='skyblue')
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)


def confusion_matrix(true, predicted, labels):
    """Computes the confusion matrix."""
    matrix = np.zeros((len(labels), len(labels)))
    for i in range(len(true)):
        matrix[true[i]][predicted[i]] += 1
    return matrix.astype(int)


def plot_confusion_matrices(true, predicted, classes, labels, ax=None):
    """Plots a confusion matrix as a heatmap."""
    matrix = confusion_matrix(true, predicted, labels)

    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 5))

    sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues", linewidths=0.5, cbar=True, ax=ax)
    ax.set_xticklabels(classes, rotation=45, ha="right")
    ax.set_yticklabels(classes, rotation=0, va="center")
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.set_title("Confusion Matrix")


def plot_class_accuracy(true, predicted, classes, labels, ax=None):
    """Plots accuracy per class as a bar chart."""
    matrix = confusion_matrix(true, predicted, labels)
    accuracy = [np.diagonal(matrix)[i] / np.sum(matrix[i]) if np.sum(matrix[i]) > 0 else 0 for i in range(len(labels))]

    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 5))

    ax.bar(classes, accuracy, color='skyblue')
    ax.set_xlabel('Class Labels')
    ax.set_ylabel('Accuracy')
    ax.set_title("Accuracy per Class")
    ax.set_xticklabels(classes, rotation=45)


def plot_true_vs_predicted_counts(true, predicted, classes, labels, ax=None):
    """Plots the number of true and predicted labels per class."""
    true_counts = Counter(true)
    predicted_counts = Counter(predicted)

    x = np.arange(len(labels))
    true_values = [true_counts.get(label, 0) for label in labels]
    pred_values = [predicted_counts.get(label, 0) for label in labels]

    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))

    width = 0.35
    ax.bar(x - width/2, true_values, width, label='True Labels', color='skyblue')
    ax.bar(x + width/2, pred_values, width, label='Predicted Labels', color='lightcoral')
    ax.set_xlabel("Class Labels")
    ax.set_ylabel("Count")
    ax.set_title("True vs Predicted Counts")
    ax.set_xticks(x)
    ax.set_xticklabels(classes, rotation=45)
    ax.legend()


def plot_side_by_side_accuracies(accuracies_attack, accuracies_no_attack, ax=None):
    """Plots side-by-side accuracy comparison."""
    langs = list(accuracies_attack.keys())
    acc_attack = [accuracies_attack[lang] for lang in langs]
    acc_no_attack = [accuracies_no_attack[lang] for lang in langs]

    x = np.arange(len(langs))
    width = 0.35

    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))

    ax.bar(x - width/2, acc_attack, width, label="With Attack", color="red")
    ax.bar(x + width/2, acc_no_attack, width, label="No Attack", color="blue")
    ax.set_xlabel("Languages")
    ax.set_ylabel("Accuracy")
    ax.set_title("Accuracy Comparison: With vs Without Attack")
    ax.set_xticks(x)
    ax.set_xticklabels(langs, rotation=45)
    ax.legend()


def plot_accuracies_and_percentage_diff(accuracies_attack, accuracies_no_attack):
    """Plots accuracy comparison and percentage difference."""
    langs = list(accuracies_attack.keys())
    acc_attack = np.array([accuracies_attack[lang] for lang in langs])
    acc_no_attack = np.array([accuracies_no_attack[lang] for lang in langs])
    percent_diff = ((acc_no_attack - acc_attack) / acc_attack) * 100  

    x = np.arange(len(langs))
    width = 0.35

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Accuracy Comparison
    axes[0].bar(x - width/2, acc_attack, width, label="With Attack", color="red")
    axes[0].bar(x + width/2, acc_no_attack, width, label="No Attack", color="blue")
    axes[0].set_xlabel("Languages")
    axes[0].set_ylabel("Accuracy")
    axes[0].set_title("Accuracy Comparison")
    axes[0].set_xticks(x)
    axes[0].set_xticklabels(langs, rotation=45)
    axes[0].legend()

    # Percentage Difference
    axes[1].bar(x, percent_diff, width, color="purple")
    axes[1].set_xlabel("Languages")
    axes[1].set_ylabel("Percentage Difference (%)")
    axes[1].set_title("Percentage Difference in Accuracy")
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(langs, rotation=45)
    axes[1].axhline(0, color="gray", linestyle="--", linewidth=0.8)

    fig.tight_layout()
    plt.show()


def plot_compare(data1, data2, langs, label1, label2, xlabel, ylabel, title, ax=None):
    """General comparison plot for two datasets."""
    x = np.arange(len(langs))
    width = 0.35

    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))

    ax.bar(x - width/2, data1, width, label=label1, color="red")
    ax.bar(x + width/2, data2, width, label=label2, color="blue")
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(langs, rotation=45)
    ax.legend()


# ===============================
# Multi-Plot Function for Each Language
# ===============================

def plot_language_analysis(lang, true, predicted, classes, labels, accuracy):
    """Plots all evaluation metrics for a single language in one figure."""
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle(f"Analysis for {lang} (No Attack)", fontsize=16)

    # Confusion Matrix
    plot_confusion_matrices(true, predicted, classes, labels, ax=axes[0, 0])
    axes[0, 0].set_title("Confusion Matrix")

    # Accuracy per Class
    plot_class_accuracy(true, predicted, classes, labels, ax=axes[0, 1])
    axes[0, 1].set_title("Accuracy per Class")

    # True vs. Predicted Counts
    plot_true_vs_predicted_counts(true, predicted, classes, labels, ax=axes[1, 0])
    axes[1, 0].set_title("True vs Predicted Counts")

    # Language Accuracy Bar Plot
    plot_histogram([accuracy], [lang], "Language Accuracy", "Language", "Accuracy", ax=axes[1, 1])
    axes[1, 1].set_title("Overall Accuracy")

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
import math

# Gini Index for a list of probabilities
def gini_index(probs):
    return 1 - sum(p**2 for p in probs)

# Entropy for a list of probabilities
def entropy(probs):
    return -sum(p * math.log2(p) for p in probs if p > 0)

# Confidence Margin for a list of probabilities
def confidence_margin(probs):
    sorted_probs = sorted(probs, reverse=True)
    return sorted_probs[0] - sorted_probs[1] if len(sorted_probs) > 1 else 0
```

```python
from collections import defaultdict

average_uncertainty = {}
languages = ['en', 'fr', 'el', 'bg', 'th', 'es']
for language in languages:
    total_gini = 0
    total_entropy = 0
    total_margin = 0
    count = 0

    for i, instance_dict in enumerate(counts[language]):
        total_count = sum(instance_dict.values())
        if total_count == 0:
            continue  # Skip empty instances

        probs = [count / total_count for count in instance_dict.values()]
        
        gini = gini_index(probs)
        ent = entropy(probs)
        margin = confidence_margin(probs)

        total_gini += gini
        total_entropy += ent
        total_margin += margin
        count += 1

    average_uncertainty[language] = {
        'gini': total_gini / count if count else 0,
        'entropy': total_entropy / count if count else 0,
        'margin': total_margin / count if count else 0
    }
print(average_uncertainty)
```
