```python
# Methods for analysis of a single run
def plot_histogram(data, labels, title, xlabel, ylabel):
    plt.bar(labels, data)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def confusion_matrix(true, predicted, labels):
    matrix = np.zeros((len(labels), len(labels)))
    for i in range(len(true)):
        matrix[true[i]][predicted[i]] += 1
    matrix = [[int(element) for element in row] for row in matrix]
    return matrix



def plot_confusion_matrices(true, predicted, classes, labels, title):
    # Compute confusion matrix using numerical labels
    matrix = confusion_matrix(true, predicted, labels=labels)

    plt.figure(figsize=(6, 5))

    # Use "Blues" colormap with annotations
    ax = sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues", linewidths=0.5, cbar=True)

    # Set class names as axis labels, ensuring they align with heatmap cells
    ax.set_xticks([i + 0.5 for i in range(len(classes))])  # Center labels
    ax.set_yticks([i + 0.5 for i in range(len(classes))])
    
    ax.set_xticklabels(classes, rotation=45, ha="right")
    ax.set_yticklabels(classes, rotation=0, va="center")

    # Labels and title
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(title)

    plt.show()
    

def plot_class_accuracy(true, predicted, classes, labels, title):
    # Calculate confusion matrix
    matrix = confusion_matrix(true, predicted, labels=labels)
    
    # Initialize an array to store accuracy for each class
    accuracy = []
    
    # Calculate accuracy for each class
    for i in range(len(labels)):
        accuracy.append(np.diagonal(matrix)[i] / np.sum(matrix[i]))  # Accuracy for each class
    
    # Plotting the accuracy per class
    x = np.arange(len(labels))  # x-axis positions for each class
    
    plt.figure(figsize=(8, 6))
    
    # Plot accuracy as a bar chart
    plt.bar(x, accuracy, color='skyblue')
    
    # Adding labels and title
    plt.xlabel('Class Labels')
    plt.ylabel('Accuracy')
    plt.title(title)
    
    # Customizing the x-ticks to show class names
    plt.xticks(x, classes, rotation=45)
    
    # Display the plot
    plt.tight_layout()
    plt.show()



def plot_true_vs_predicted_counts(true, predicted, classes, labels, title):
    # Count the number of occurrences of each class in true and predicted labels
    true_counts = Counter(true)
    predicted_counts = Counter(predicted)
    
    # Prepare data for plotting
    x = np.arange(len(labels))  # x-axis positions for each class
    true_counts_values = [true_counts.get(label, 0) for label in labels]  # True label counts
    predicted_counts_values = [predicted_counts.get(label, 0) for label in labels]  # Predicted label counts
    
    # Plotting the number of true and predicted labels per class side by side
    width = 0.35  # Width of the bars
    
    plt.figure(figsize=(10, 6))
    
    # Plotting the bars for true counts and predicted counts
    plt.bar(x - width/2, true_counts_values, width, label='True Labels', color='skyblue')
    plt.bar(x + width/2, predicted_counts_values, width, label='Predicted Labels', color='lightcoral')
    
    # Adding labels and title
    plt.xlabel('Class Labels')
    plt.ylabel('Count')
    plt.title(title)
    
    # Customizing the x-ticks to show class names
    plt.xticks(x, classes, rotation=45)
    
    # Adding legend
    plt.legend()
    
    # Display the plot
    plt.tight_layout()
    plt.show()
    
def filter_none_three(arr1, arr2, arr3):
    """
    Removes elements where any of the three arrays has a None value.

    Parameters:
    - arr1: list of values (can include None)
    - arr2: list of values (can include None)
    - arr3: list of values (can include None)

    Returns:
    - Tuple of three lists with None-containing elements removed.
    """
    if len(arr1) != len(arr2) or len(arr1) != len(arr3):
        raise ValueError("All arrays must have the same length.")

    filtered_arr1, filtered_arr2, filtered_arr3 = [], [], []

    for a, b, c in zip(arr1, arr2, arr3):
        if a is not None and b is not None and c is not None:
            filtered_arr1.append(a)
            filtered_arr2.append(b)
            filtered_arr3.append(c)

    return filtered_arr1, filtered_arr2, filtered_arr3

def plot_side_by_side_accuracies(accuracies_attack, accuracies_no_attack, label1 = "With Attack", label2 = "No Attack", title = "Accuracy Comparison: With vs Without Attack"):
    """
    Plots a side-by-side bar chart comparing accuracy with and without attack.

    Parameters:
    - accuracies_attack: dict {language: accuracy} for the attack scenario
    - accuracies_no_attack: dict {language: accuracy} for the no-attack scenario
    """
    langs = list(accuracies_attack.keys())  # Extract language labels
    acc_attack = [accuracies_attack[lang] for lang in langs]
    acc_no_attack = [accuracies_no_attack[lang] for lang in langs]

    x = np.arange(len(langs))  # Positions for bars
    width = 0.35  # Bar width for side-by-side plotting

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.bar(x - width/2, acc_attack, width, label=label1, color="red")
    ax.bar(x + width/2, acc_no_attack, width, label=label2, color="blue")

    # Labels and formatting
    ax.set_xlabel("Languages")
    ax.set_ylabel("Accuracy")
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(langs, rotation=45)
    ax.legend()

    plt.tight_layout()
    plt.show()
    

def plot_accuracies_and_percentage_diff(accuracies_attack, accuracies_no_attack, label1 = "With Attack", label2 = "No Attack"):
    """
    Plots two bar charts:
    1. Side-by-side accuracy comparison (with attack vs no attack).
    2. Percentage difference in accuracy.

    Parameters:
    - accuracies_attack: dict {language: accuracy} for the attack scenario
    - accuracies_no_attack: dict {language: accuracy} for the no-attack scenario
    """
    langs = list(accuracies_attack.keys())  # Extract language labels
    acc_attack = np.array([accuracies_attack[lang] for lang in langs])
    acc_no_attack = np.array([accuracies_no_attack[lang] for lang in langs])

    # Compute percentage difference
    percent_diff = ((acc_no_attack - acc_attack) / acc_attack) * 100  

    x = np.arange(len(langs))  # X-axis positions
    width = 0.35  # Bar width for side-by-side plotting

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))  # Two subplots side by side

    # First subplot: Accuracy comparison
    axes[0].bar(x - width/2, acc_attack, width, label=label1, color="red")
    axes[0].bar(x + width/2, acc_no_attack, width, label=label2, color="blue")

    axes[0].set_xlabel("Languages")
    axes[0].set_ylabel("Accuracy")
    axes[0].set_title("Accuracy Comparison")
    axes[0].set_xticks(x)
    axes[0].set_xticklabels(langs, rotation=45)
    axes[0].legend()

    # Second subplot: Percentage difference
    axes[1].bar(x, percent_diff, width, color="purple")
    axes[1].set_xlabel("Languages")
    axes[1].set_ylabel("Percentage Difference (%)")
    axes[1].set_title("Percentage Difference in Accuracy")
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(langs, rotation=45)
    axes[1].axhline(0, color="gray", linestyle="--", linewidth=0.8)  # Reference line at 0%

    fig.tight_layout()
    plt.show()
```
