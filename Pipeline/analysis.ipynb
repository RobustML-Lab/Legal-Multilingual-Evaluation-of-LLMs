```python
import matplotlib.pyplot as plt
from data import *
```

```python
dataset = Dataset.get_dataset('multi_eurlex')
data, label_options = dataset.get_data('en')
```

```python
label_distribution = {i: 0 for i, label in enumerate(label_options)}
average_labels = 0
for item in data:
    labels = item['labels']
    average_labels += len(labels)
    for label in labels:
        label_distribution[label] += 1
average_labels = average_labels/len(data)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(label_options, label_distribution.values(), color='skyblue')
plt.title("Distribution of True Labels")
plt.xlabel('Category')
plt.ylabel('True Num')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
print("Average number of labels per documents: ", average_labels)
```
